```python
import control as ctl
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
s = ctl.TransferFunction.s
T = 0.001
time = np.arange(0, 0.3, T)

R = 15
L = 0.047
C = 200*10**(-6)

G = 1/(C*L*s**2 + C*R*s + 1)
display(G)

time, response = ctl.step_response(G)
plt.plot(time, response)

time, response = ctl.step_response(G, T=np.arange(0, time[-1], T))
plt.scatter(time, response, s=5)
plt.grid()
```
